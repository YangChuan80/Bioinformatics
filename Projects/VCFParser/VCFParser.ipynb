```python
import numpy as np
import pandas as pd
```

```python
def loadVCF(filename):
    f = open(filename)
    variants = []
    
    while 1:
        line = f.readline().rstrip()
        if len(line) == 0:
            break
        else:
            if line[0:2] == '##':
                continue
            elif line[0] == '#':
                header = line[1:].split('\t')
            else:
                variant = line.split('\t')
                variants.append(variant)
    df = pd.DataFrame(variants, columns = header)
    return df
```

```python
df = loadVCF('example.vcf')
```

```python
df.to_csv('check.csv')
```
