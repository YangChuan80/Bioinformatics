## Parsing Structure Files

```python
parser = Bio.PDB.MMCIFParser()
```

```python
structure = parser.get_structure('SpikeProteinSARS', 'wr/5wrg.cif')
```

```python
for i in structure.get_chains():
    print(i)
```
