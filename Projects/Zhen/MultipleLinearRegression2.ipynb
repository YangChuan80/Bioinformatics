## Import Statistical Module

```python
import xlrd
import pandas as pd
import numpy as np
import sys
from scipy import stats
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load the File

```python
#Load the File
filename='.\\BK for Wei 2 from Zhen.xlsx'
book=xlrd.open_workbook(filename)
print('Source file: '+sys.path[0]+filename+' loaded!')
```

```python
nrows3 = book.sheet_by_name('BK_UA').nrows
nrows2 = book.sheet_by_name('BK_CRP').nrows
nrows1 = book.sheet_by_name('BK_lipid').nrows
```

```python
header3 = book.sheet_by_name('BK_UA').row_values(0)
header2 = book.sheet_by_name('BK_CRP').row_values(0)
header1 = book.sheet_by_name('BK_lipid').row_values(0)
```

```python
data3 = [book.sheet_by_name('BK_UA').row_values(i) for i in range(1, nrows3)]
data2 = [book.sheet_by_name('BK_CRP').row_values(i) for i in range(1, nrows2)]
data1 = [book.sheet_by_name('BK_lipid').row_values(i) for i in range(1, nrows1)]
```

```python
df3 = pd.DataFrame(data3, columns = header3)
df2 = pd.DataFrame(data2, columns = header2)
df1 = pd.DataFrame(data1, columns = header1)
```

## Extract Raw DataFrame

```python
df2
```

## Drop NA Data

```python
df3_dropna = df3.dropna()
df2_dropna = df2.dropna()
df1_dropna = df1.dropna()
```

## Drop Angiotensin Converting Enzyme Inhibitor (ACEI) Positive Data

```python
df3_withoutACEI = df3_dropna[df3_dropna['ACE inhibitor'] != 1]
df2_withoutACEI = df2_dropna[df2_dropna['ACE inhibitor'] != 1]
df1_withoutACEI = df1_dropna[df1_dropna['ACE inhibitor'] != 1]
```

```python
df2_withoutACEI
```

## Multiple Linear Regression

```python
parameters3 = header3[6:]
parameters2 = header2[6:]
parameters1 = header1[6:]
```

```python
data3 = df3_withoutACEI[parameters3]
data2 = df2_withoutACEI[parameters2]
data1 = df1_withoutACEI[parameters1]
```

```python
data3 = data3.astype(float)
data2 = data2.astype(float)
data1 = data1.astype(float)
```

```python
data3
```

```python
para_str3 = ''
for parameter in parameters3:
    para_str3 = para_str3 + '+' + parameter
    
para_str2 = ''
for parameter in parameters2:
    para_str2 = para_str2 + '+' + parameter
    
para_str1 = ''
for parameter in parameters1:
    para_str1 = para_str1 + '+' + parameter
```

```python
para_str3 = para_str3[17:]
para_str3 = 'HypBKBK~'+para_str3

para_str2 = para_str2[17:]
para_str2 = 'HypBKBK~'+para_str2

para_str1 = para_str1[17:]
para_str1 = 'HypBKBK~'+para_str1

para_str1
```

### HypBK/BK vs Other Parameters

```python
model3 = ols(para_str3, data3).fit()
model3.summary()
```

```python
model2 = ols(para_str2, data2).fit()
model2.summary()
```

```python
model1 = ols(para_str1, data1).fit()
model1.summary()
```

### BK/HypBK vs Other Parameters

```python
para_str3_2 = para_str3[8:]
para_str3_2 = 'BKHypBK~'+para_str3_2

para_str2_2 = para_str2[8:]
para_str2_2 = 'BKHypBK~'+para_str2_2

para_str1_2 = para_str1[8:]
para_str1_2 = 'BKHypBK~'+para_str1_2

para_str3_2
```

```python
model3_2 = ols(para_str3_2, data3).fit()
model3_2.summary()
```

```python
model2_2 = ols(para_str2_2, data2).fit()
model2_2.summary()
```
