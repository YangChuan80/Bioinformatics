## Import Statistical Module

```python
import xlrd
import pandas as pd
import numpy as np
import sys
from scipy import stats
```

## Load the File

```python
#Load the File
filename='.\\BK for Wei.xlsx'
book=xlrd.open_workbook(filename)
print('Source file: '+sys.path[0]+filename+' loaded!')
```

```python
nrows = book.sheet_by_name('Sheet1').nrows
```

```python
nrows
```

```python
header = book.sheet_by_name('Sheet1').row_values(0)
```

```python
data = [book.sheet_by_name('Sheet1').row_values(i) for i in range(1, nrows)]
```

```python
df = pd.DataFrame(data, columns = header)
```

## Extract Raw DataFrame

```python
df
```

## Drop NA Data

```python
for parameter in header[11:-1]:
    df[df[parameter] == ''] = np.nan
    
for parameter in header[8:10]:
    df[df[parameter] == ''] = np.nan    

df[df['BMI'] == ''] = np.nan
```

```python
df_dropna = df.dropna()
```

## Drop Angiotensin Converting Enzyme Inhibitor (ACEI) Positive Data

```python
df_withoutACEI = df_dropna[df_dropna['ACE inhibitor'] != 1]
```

```python
df_withoutACEI
```

## Pearson's Correlation Between HypBK/BK vs other parameters

```python
result = stats.pearsonr(df_withoutACEI['HypBK/BK'].values, df_withoutACEI['BMI'].values)
print('HypBK/BK vs ' + 'BMI' + ': ')
print('Pearson’s correlation coefficient = ', result[0])
print('p-value = ', result[1])
print()
    
for parameter in header[10:]:
    print(parameter)
    result = stats.pearsonr(df_withoutACEI['HypBK/BK'].values, df_withoutACEI[parameter].values)
    print('HypBK/BK vs ' + parameter + ': ')
    print('Pearson’s correlation coefficient = ', result[0])
    print('p-value = ', result[1])
    print()
```

## Pearson's Correlation Between BK/HypBK vs other parameters

```python
result = stats.pearsonr(df_withoutACEI['BK/ HypBK'].values, df_withoutACEI['BMI'].values)
print('BK/ HypBK vs ' + 'BMI' + ': ')
print('Pearson’s correlation coefficient = ', result[0])
print('p-value = ', result[1])
print()
    
for parameter in header[10:]:
    print(parameter)
    result = stats.pearsonr(df_withoutACEI['BK/ HypBK'].values, df_withoutACEI[parameter].values)
    print('BK/ HypBK vs ' + parameter + ': ')
    print('Pearson’s correlation coefficient = ', result[0])
    print('p-value = ', result[1])
    print()
```
