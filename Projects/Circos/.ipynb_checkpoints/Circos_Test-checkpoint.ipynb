```python
from circos import CircosPlot
%matplotlib inline
```

```python
nodes = []
for i in range(100):
    nodes.append(i)
```

```python
edges = [(0,40), (20, 30), (0,15), (0,60), (0,80)]
```

```python
#nodeprops = dict(facecolor='green', alpha=1)
#edgeprops = dict(edgecolor='red', alpha=1, ls='--')
```

```python
c = CircosPlot(nodes, edges, radius = 10, 
              nodecolor="green",
              edgecolor="red",
              )
c.draw()
```
