```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
%matplotlib inline
```

```python
image = Image.open('Lead_II.bmp')
image
```

```python
x = image.size[0]
y = image.size[1]
```

```python
print(x)
print(y)
```

```python
matrix = []
points = []
integrated_density = 0

for i in range(x):
    matrix.append([])
    for j in range(y):
        matrix[i].append(image.getpixel((i,j)))
        #integrated_density += image.getpixel((i,j))[1]
        #points.append(image.getpixel((i,j))[1])
```

### Extract Blue Points

```python
redline_pos = 0
gain = 1
absMax = 0
II = []
points_vertical = []
theOne = 0

for xAxis in range(x):
    for yAxis in range(y):
        currentPoint = matrix[xAxis][yAxis]
        # Pickup Blue points
        if currentPoint[0] < 100 and currentPoint[1] < 100 and currentPoint[2] < 100:
            points_vertical.append(yAxis)
            
    print(points_vertical)
    
    
    # Choose the largest amplitude
    for item in points_vertical:
        
        if math.fabs(item-redline_pos) > absMax:
            absMax = math.fabs(item-redline_pos)
            theOne = item 
    II.append((200-theOne)*gain)
    
    absMax = 0  
    theOne = 0
    points_vertical = []
```
